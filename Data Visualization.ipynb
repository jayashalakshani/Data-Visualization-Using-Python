```python
df=pd.read_csv("winequality-red.csv")
column_pairs=[('alcohol','pH'),('fixed acidity','citric acid')]
fig,axs=plt.subplots(1,2,figsize=(10,5))
axs[0].scatter(df[column_pairs[0][0]],df[column_pairs[0][1]],color='purple')
axs[0].set_xlabel(column_pairs[0][0])
axs[0].set_ylabel(column_pairs[0][1])

axs[1].scatter(df[column_pairs[1][0]],df[column_pairs[1][1]],color='red')
axs[1].set_xlabel(column_pairs[1][0])
axs[1].set_ylabel(column_pairs[1][1])

plt.show()
```

```python
# share the x and y axis.
df=pd.read_csv("winequality-red.csv")
column_pairs=[('alcohol','pH'),('fixed acidity','citric acid')]
fig,axs=plt.subplots(1,2,figsize=(10,5),sharex=True,sharey=True)
axs[0].scatter(df[column_pairs[0][0]],df[column_pairs[0][1]],color='purple')
axs[0].set_xlabel(column_pairs[0][0])
axs[0].set_ylabel(column_pairs[0][1])

axs[1].scatter(df[column_pairs[1][0]],df[column_pairs[1][1]],color='red')
axs[1].set_xlabel(column_pairs[1][0])
axs[1].set_ylabel(column_pairs[1][1])

plt.show()
```

```python
# Histrograms
titanic_data=pd.read_csv("titanic.csv")
plt.hist(titanic_data['Age'],bins=80,color='purple')
plt.xlabel('Age')
plt.ylabel('Count')
plt.title("Histrogram of Titanic Passengers Age")
plt.show()
```

```python
# Heat maps
df=pd.read_csv("winequality-red.csv")
# calculate the correlation matrix
corr_matrix=df.corr()
# create heapmap
plt.figure(figsize=(10,8))
plt.imshow(corr_matrix,cmap='coolwarm',aspect='auto')
plt.colorbar(format='%.2f')
plt.title('Correlation Heatmap')
plt.xticks(range(len(corr_matrix)), corr_matrix.columns, rotation=90)
plt.yticks(range(len(corr_matrix)), corr_matrix.columns)
plt.tight_layout()

plt.show()
```

```python
# Data visualization using pandas library
df = pd.read_csv("mpg.csv")
df.head()
```

```python
# line plot
plt.figure()
df.groupby('model year')['mpg'].mean().plot(kind='line')
plt.xlabel("Model Year")
plt.ylabel('Mean MPG')
plt.title("Mean MPG by Model years")
plt.show()
```

```python
# Scatter plot
plt.figure()
df.plot(kind='scatter',x='displacement',y='mpg')
plt.xlabel('Displacement')
plt.ylabel('MPG')
plt.title('Fule efficiency by Displacement')
# save the figure
plt.savefig('scatterplt.png')
plt.show()
```

```python
# Bar plot
plt.figure()
df.groupby('origin')['mpg'].mean().plot(kind='bar')
plt.xlabel("Origin")
plt.ylabel('Mean MPG')
plt.title("Mean MPG by Origin")
plt.show()
```

```python
# Histrograms
plt.figure()
df['acceleration'].plot(kind='hist',bins=30)
plt.xlabel("Acceleration")
plt.ylabel('Count')
plt.title("Distribution of the Acceleration")
plt.show()
```

```python
# boxplot
plt.figure()
df.boxplot(column='mpg', by='cylinders')
plt.title("Box plot")
plt.show()
```

```python
#Area plot
plt.figure()
df.groupby('model year')['mpg'].mean().plot(kind='area',alpha=0.5)
plt.xlabel("Model year")
plt.ylabel('Mean MPG')
plt.title("Mean MPG by model year")
plt.show()
```

```python
# Pie chart
df['origin'].value_counts().plot(kind='pie',colors=['purple','yellow','green'])
plt.title("Pie chart")
plt.show()
```

```python
# Kernel density estimation plots(KDE)
plt.figure()
df['mpg'].plot(kind='kde')
plt.title('KDE of mpg')
plt.show()
```

```python
df=sns.load_dataset('penguins')
df.head()
```

```python
# Scatter plot
sns.scatterplot(x='bill_length_mm',y='bill_depth_mm',data= df,marker='^')
sns.set_style('dark')
plt.show()
```

```python
# Line plot
sns.lineplot(x='body_mass_g',y='flipper_length_mm',data=df,hue='species')
plt.show()
```

```python
# Histrogram
sns.histplot(x='body_mass_g',data=df,color='green',bins=15)
plt.show()
```

```python
# boxplot
sns.boxplot(x='species',y='body_mass_g',data=df)
plt.show()
```

```python
# violin plot
sns.violinplot(x='species',y='body_mass_g',data=df)
plt.show()
```

```python
# Heatmap
corr_matrix=df.corr()
sns.heatmap(corr_matrix,annot=True)
plt.show()
```

```python
# pair plot
sns.pairplot(data=df,hue='species')
plt.show()
```

```python
# bar plot
sns.barplot(x='species',y='body_mass_g',data=df)
plt.show()
```

```python
# count plot
sns.countplot(x='species' ,data=df)
plt.show()
```
